```python
a = np.array([1,0,0,0,0,0,0,1, 3, 1, 4, 0, 0, 0])
changes_idx = (a[1:]-a[:-1]).nonzero()[0]
changes_idx
```

```python
for i in range(changes_idx.shape[0]):
    if changes_idx[i] == 0:
        val = a[changes_idx[i]]
        count = i + 1
    else:
        val = a[changes_idx[i]]
        count = changes_idx[i] - changes_idx[i-1]

    print(val, count)
```

```python
count = 0
new_el = True
res = []
for i in range(a.shape[0]):
    val = a[i]
    if val != 0:
        res.append(val)

    else:
        if new_el:
            res.append(val)
            new_el = False
            count = 0
        count += 1
        if i + 1 == a.shape[0] or a[i+1] != 0:
            res.append(count)
            new_el = True
res
```
